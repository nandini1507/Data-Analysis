```python
### Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Load Dataset
df = pd.read_csv("sales.csv")
df.head()
```

```python
#exploration
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nSummary stats:")
print(df.describe())
df.isnull().sum()
```

```python
#grouping
sales_by_product = df.groupby("Product")["Sales"].sum()
print(sales_by_product)
sales_by_region = df.groupby("Region")["Sales"].sum()
print(sales_by_region)
```

```python
#filtering examples
high_sales = df[df["Sales"] > 1000]
high_sales
```

```python
### Visualizations
sales_by_product.plot(kind="bar", title="Total Sales by Product", color="skyblue")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_by_region.plot(kind="bar", title="Total Sales by Region", color="orange")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.plot(x="Date", y="Sales", kind="line", title="Sales Trend Over Time", marker="o")
plt.show()
```
